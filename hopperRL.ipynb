```python
import gymnasium as gym
from stable_baselines3 import PPO
import os

# 1. ë¡œë´‡ í™˜ê²½ ë¶ˆëŸ¬ì˜¤ê¸° (Hopper-v4: ì™¸ë°œ ì í”„ ë¡œë´‡)
# ì´ë¯¸ ì¡°ëª…, ë°”ë‹¥, ë¬¼ë¦¬ ë²•ì¹™ì´ ì™„ë²½í•˜ê²Œ ì„¸íŒ…ë˜ì–´ ìˆìŠµë‹ˆë‹¤.
env = gym.make("Hopper-v4", render_mode="rgb_array")

# 2. ì¸ê³µì§€ëŠ¥ ëª¨ë¸(ë‡Œ) ìƒì„±

from stable_baselines3.common.torch_layers import BaseFeaturesExtractor
import torch.nn as nn

policy_kwargs = dict(activation_fn=nn.Tanh, net_arch=dict(pi=[256, 256], vf=[256, 256]))

model = PPO("MlpPolicy", env, ent_coef=0.01, policy_kwargs=policy_kwargs, verbose=1)

# 3. í•™ìŠµ ì‹œì‘! (ì•½ 3~5ë¶„ ì†Œìš”)
# 50,000ë²ˆ ì •ë„ ì—°ìŠµí•˜ë©´ ë¹„í‹€ê±°ë¦¬ë©° ê±·ê¸° ì‹œì‘í•©ë‹ˆë‹¤.
print("ğŸ¤– ë¡œë´‡ì´ ê±¸ìŒë§ˆë¥¼ ë°°ìš°ëŠ” ì¤‘ì…ë‹ˆë‹¤... (ì ì‹œë§Œ ê¸°ë‹¤ë ¤ì£¼ì„¸ìš”)")
model.learn(total_timesteps=200000)

# 4. í•™ìŠµëœ ë‡Œ ì €ì¥í•˜ê¸°
model.save("my_hopper_robot")
print("âœ… í•™ìŠµ ì™„ë£Œ! ëª¨ë¸ì´ êµ¬ê¸€ ë“œë¼ì´ë¸Œì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
from gymnasium.wrappers import RecordVideo
from IPython.display import HTML
from base64 import b64encode

# 1. ì €ì¥ëœ ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
model = PPO.load("my_hopper_robot")

# 2. ì˜ìƒì„ ì°ì„ í™˜ê²½ ë‹¤ì‹œ ë§Œë“¤ê¸°
video_folder = "videos"
env = gym.make("Hopper-v4", render_mode="rgb_array")
# RecordVideo: ë¡œë´‡ì˜ ì›€ì§ì„ì„ ë…¹í™”í•´ì£¼ëŠ” ë„êµ¬ (ê°€ì¥ ê¸´ ì—í”¼ì†Œë“œ í•˜ë‚˜ë§Œ ë…¹í™”)
env = RecordVideo(env, video_folder=video_folder, episode_trigger=lambda x: x == 0)

# 3. ë¡œë´‡ ì›€ì§ì—¬ë³´ê¸° (í…ŒìŠ¤íŠ¸)
obs, _ = env.reset()
done = False
total_reward = 0

while not done:
    action, _ = model.predict(obs) # ëª¨ë¸ì´ í–‰ë™ ê²°ì •
    obs, reward, terminated, truncated, info = env.step(action)
    total_reward += reward
    done = terminated or truncated

env.close()
print(f"ğŸ† íšë“ ì ìˆ˜: {total_reward:.2f}")

# 4. ì˜ìƒ ì¬ìƒ í•¨ìˆ˜
def show_video():
    mp4_list = [f for f in os.listdir(video_folder) if f.endswith('.mp4')]
    if len(mp4_list) > 0:
        mp4 = os.path.join(video_folder, mp4_list[0])
        video = open(mp4, "rb").read()
        data_url = "data:video/mp4;base64," + b64encode(video).decode()
        return HTML(f"""
        <div style="display:flex; justify-content:center;">
            <video width=500 controls autoplay loop style="border: 2px solid #333; border-radius: 10px;">
                <source src="{data_url}" type="video/mp4">
            </video>
        </div>
        """)
    else:
        return "âŒ ë…¹í™”ëœ ì˜ìƒì´ ì—†ìŠµë‹ˆë‹¤."

show_video()
```
